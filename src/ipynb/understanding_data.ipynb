# アイドルIDの一覧DataFrame作る

先輩や同期などは省いて、アイドルのみのツイートを持ってきたい。アイドルIDの一覧DataFrameを作って、そこと照らし合わせながらテキスト持ってきたい。

```python
result = pd.DataFrame(columns=['id','screen_name'])
```

```python
for index, row in base.iterrows():
    name = row[0]['user']['screen_name']
    id = row[0]['user']['id']
    print(f'ID：{id}')

    result = result.append({ 'id':id,'screen_name':name},ignore_index=True)

result = result[~result.duplicated(subset='id',keep='last')]
```
